```python
# Identification des donnes aberrantes - IQR
# Fonction pour dtecter les outliers en utilisant l'IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Trouver les outliers pour la colonne 'PRICE'
outliers = detect_outliers_iqr(df,'cigsPerDay')
# Afficher les outliers
print("Outliers dtects:")
print(outliers)
print(len(outliers))
```

**On remarque que cette variable contient 12 valeurs aberrntes egale a 60,70 cigarettes per day chose illogique qui peut etre due a une faute de frappe donc on va essayer de remplacer ces valeurs pour la moyenne au lieu de les supprimer**

```python
def replace_outliers_with_mean(data, column):
    # Calcul des quartiles et des bornes
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Calcul de la moyenne de la colonne (sans tenir compte des NaN)
    column_mean = data[column].mean()
    
    # Remplacement des valeurs aberrantes
    data[column] = data[column].apply(lambda x: column_mean if (x < lower_bound or x > upper_bound) else x)
    return data

# Application de la fonction  la colonne 'cigsPerDay'
df = replace_outliers_with_mean(df, 'cigsPerDay')

# Vrification
print("Colonnes aprs traitement des outliers :")
print(df['cigsPerDay'].describe())
```

```python


# Analyse univarie pour une variable quantitative
def analyze_quantitative(df, column):
    print("Statistiques descriptives :")
    print(df[column].describe())
    sns.kdeplot(data=df, x=column, fill=True, color="purple", bw_adjust=0.5)
    plt.title("Fonction de densit")
    plt.show()

```

**Sexe**

```python
# Analyse univarie pour une variable qualitative
def analyze_categorical(df, column):
    # Frquences absolues
    frequences = df[column].value_counts()
    print("Frquences absolues :\n", frequences)

    # Frquences relatives
    proportions = df[column].value_counts(normalize=True) * 100
    print("\nProportions (%) :\n", proportions)

    # Mode et catgories uniques
    mode = df[column].mode()[0]
    n_categories = df[column].nunique()
    print("\nMode :", mode)
    print("\nNombre de catgories uniques :", n_categories)
    print("\nCatgories uniques :", df[column].unique())

    # Visualisation
    print("\nVisualisation :")
    df[column].value_counts().plot(kind='bar', color='orange')
    plt.title(f'Distribution de {column}')
    plt.xlabel('Catgorie')
    plt.ylabel('Frquence')
    plt.show()


```

**Feature Selection**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

X = df.drop(columns=['TenYearCHD'])  
y = df['TenYearCHD']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = LogisticRegression(solver='liblinear', max_iter=1000)

# Forward Selection
sfs = SFS(clf, k_features='best', forward=True, floating=False, scoring='accuracy', cv=5)
sfs = sfs.fit(X_train, y_train)

# Correction pour afficher les noms des caractristiques slectionnes
print("Forward Selection: \nMeilleures caractristiques:", list(X.columns[list(sfs.k_feature_idx_)]),
      "\nMeilleur score:", sfs.k_score_)
```

**Interprtation du modle**

```python
from imblearn.over_sampling import SMOTE


X = df.drop(columns=['TenYearCHD'])  
y = df['TenYearCHD']
smote=SMOTE(sampling_strategy='minority') 
x,y=smote.fit_resample(X,y)
y.value_counts()
XX=x[['male', 'age', 'currentSmoker', 'prevalentStroke', 'prevalentHyp', 'diabetes', 'cigsPerDay_imputed', 'sysBP_imputed', 'glucose_imputed', 'BPMeds_imputed'] ]
```

```python
import statsmodels.api as sm



X_train, X_test, y_train, y_test = train_test_split(XX, y, test_size=0.3, random_state=42)

# Crer et ajuster le modle de rgression logistique
model = sm.Logit(y_train, X_train)
result = model.fit()

# Afficher le rsum du modle
print(result.summary())
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Prdictions sur l'ensemble de test
y_pred = result.predict(X_test)  # Prdictions des probabilits
y_pred_class = (y_pred > 0.5).astype(int)  # Conversion des probabilits en classes (0 ou 1)



cm1 = confusion_matrix(y_test, y_pred_class)
ConfusionMatrixDisplay(cm1).plot()



# valuation du modle
accuracy = accuracy_score(y_test, y_pred_class)
class_report = classification_report(y_test, y_pred_class)

# Affichage des rsultats
print(f"Accuracy: {accuracy:.2f}")
print("Rapport de classification:")
print("Matrice de confusion:")
print(class_report)
```

```python
from sklearn.metrics import roc_curve, auc

# Calcul des taux de vrais positifs (TPR) et de faux positifs (FPR)
y_pred_proba = result.predict(X_test)
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Calcul de l'AUC
roc_auc = auc(fpr, tpr)

# Affichage de la courbe ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='Courbe ROC (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taux de faux positifs (FPR)')
plt.ylabel('Taux de vrais positifs (TPR)')
plt.title('Courbe ROC')
plt.legend(loc='lower right')
plt.show()

# Affichage de l'AUC
print(f"AUC: {roc_auc}")
```
